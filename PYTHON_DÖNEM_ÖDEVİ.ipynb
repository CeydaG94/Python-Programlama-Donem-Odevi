```python
# Fonksiyon
def tutar_hesapla(aylik_ucret):
    """
    Parametre olarak gelen ücrete %20 KDV ekler.
    """
    kdv_dahil_tutar = aylik_ucret * 1.20
   # %20 eklemek demek 1.20 ile çarpmak demektir
    return kdv_dahil_tutar

# Fonksiyon testi  (Örn: 100 TL için sonuç 120.0 olmalı)
test_tutari = 100
sonuc = tutar_hesapla(test_tutari)
print(f"Test Sonucu: 100 TL -> {sonuc} TL (KDV Dahil)")
```

```python
import math      # Yuvarlama işlemleri için
import datetime  # Tarih bilgisi için

# Örnek bir müşteri ücreti alalım (Önceki listeden veya rastgele)
orijinal_ucret = 550.45

# 1. Daha önce yazdığımız fonksiyonu kullanarak KDV'li tutarı hesaplayalım
kdv_li_tutar = tutar_hesapla(orijinal_ucret) # Sonuç örn: 660.54

# 2. math kütüphanesi ile YUKARI yuvarlama
# math.ceil() fonksiyonu sayıyı ne olursa olsun bir üst tam sayıya tamamlar.
toplam_fatura = math.ceil(kdv_li_tutar)

# 3. datetime kütüphanesi ile bugünün tarihini alma
bugun = datetime.datetime.now()
tarih_formati = bugun.strftime("%d/%m/%Y") # Gün/Ay/Yıl formatına çevirdik

# Sonuçları Yazdıralım
print(f"KDV Dahil Ham Tutar: {kdv_li_tutar:.2f} TL")
print(f"Yuvarlanmış Toplam Fatura: {toplam_fatura} TL")
print("-" * 30)
print(f"Fatura Tarihi: {tarih_formati}")
```

```python
# AI TARAFINDAN HAZIRLANMIŞ VE DERLENMİŞ KOD #
import random    # Kütüphane Kullanımı (Rastgele Sayı)
import math      # Kütüphane Kullanımı (Matematiksel İşlemler)
import datetime  # Kütüphane Kullanımı (Zaman ve Tarih)

# --- 1. KISIM: TEMEL MANTIK ---

# Değişken Tanımlama ve Veri Tipleri (String, Float, Int, Boolean)
ad_soyad = "Ahmet Yılmaz"
aylik_ucret = 550.0
sadakat_ayi = 30
aktif_mi = True

# Liste Kullanımı (Veri Yapıları)
hizmetler = ["İnternet", "Televizyon", "Mobil Veri", "Sabit Telefon", "Bulut Depolama"]

# Sözlük (Dictionary) Kullanımı (Veri Yapıları)
musteri_verisi = {
    "ad_soyad": ad_soyad,
    "aylik_ucret": aylik_ucret,
    "sadakat_ayi": sadakat_ayi,
    "aktif_mi": aktif_mi
}

# Koşullu İfadeler ve Mantıksal Operatörler (If-Else / Or)
if musteri_verisi["aylik_ucret"] > 500 or musteri_verisi["sadakat_ayi"] > 24:
    status = "VIP Müşteri: İndirim Tanımla" # Koşul sağlanırsa çalışır
else:
    status = "Standart Müşteri" # Koşul sağlanmazsa çalışır

# String Metotları (Büyük Harfe Çevirme)
ad_buyuk = musteri_verisi["ad_soyad"].upper()

# Random Kütüphanesi ve String Formatlama (f-string)
customerID = f"IST-2026-{random.randint(1000, 9999)}"

# --- 2. KISIM: İLERİ SEVİYE YAPI ---

# Liste İçinde Sözlük Yapısı (Karmaşık Veri Yapıları)
musteriler_listesi = [
    {"ad": "Ahmet", "ucret": 550, "sure": 30},
    {"ad": "Ayşe", "ucret": 420, "sure": 12},
    {"ad": "Mehmet", "ucret": 600, "sure": 45}
]

# Fonksiyon Tanımlama (Parametre ve Return Kullanımı)
def tutar_hesapla(ucret):
    return ucret * 1.20 # %20 KDV ekleyip geri döndürme

# math Kütüphanesi Kullanımı (Yukarı Yuvarlama)
toplam_fatura = math.ceil(tutar_hesapla(musteri_verisi["aylik_ucret"]))

# datetime Kütüphanesi Kullanımı (Tarih Formatlama)
tarih = datetime.datetime.now().strftime("%d-%m-%Y")

# Set (Küme) Kullanımı (Benzersiz Eleman Bulma / Veri Temizleme)
tekrarlı_hizmetler = ["İnternet", "TV", "İnternet", "Mobil"]
benzersiz_hizmetler = set(tekrarlı_hizmetler)

# Çıktı Alma (Ekran Yazdırma)
print(f"ID: {customerID} | Müşteri: {ad_buyuk} | Durum: {status}")
print(f"Fatura Tutarı (KDV Dahil): {toplam_fatura} TL | Tarih: {tarih}")
print(f"Hizmet Çeşitleri: {benzersiz_hizmetler}")
```
